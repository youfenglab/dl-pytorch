```python
import torch
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
def plot_fun(X, y, fun_str):
    plt.figure(figsize = (5, 2.5))
    plt.plot(X.detach(), y.detach())
    plt.title(str(fun_str))
    plt.xlabel("X")
    plt.ylabel(str(fun_str) + "(X)")
    plt.grid("True")
    plt.show()
    
def plot_grad(X, fun_str):
    plt.figure(figsize = (5, 2.5))
    plt.plot(X.detach(), X.grad)
    plt.title(str(fun_str))
    plt.xlabel("X")
    plt.ylabel("grad of " + str(fun_str))
    plt.grid("True")
    plt.show()
```

```python
# relu function
X = torch.arange(-8.0, 8.0, 0.1, requires_grad=True)
y = torch.relu(X)
plot_fun(X, y, "relu")
```

```python
y.backward(torch.ones_like(X), retain_graph=True)
plot_grad(X, "relu")
```

```python
# sigmoid function
y = torch.sigmoid(X)
plot_fun(X, y, "sigmoid")
```

```python
X.grad.data.zero_()
y.backward(torch.ones_like(X), retain_graph=True)
plot_grad(X, "sigmoid")
```

```python
# tanh
y = torch.tanh(X)
plot_fun(X, y, "tanh")
```

```python
X.grad.data.zero_()
y.backward(torch.ones_like(X), retain_graph=True)
plot_grad(X, "tanh")
```
